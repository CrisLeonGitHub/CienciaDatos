#### Identificar y procesar los datos omitidos

Convertir "?" a NaN

```python

import numpy as np
# replace "?" to NaN
df.replace("?", np.nan, inplace = True)
df.head(5)
```

Evaluar los Datos Omitidos

```python
missing_data = df.isnull()
missing_data.head(5)
```

Contar los valores omitidos en cada columna

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

#### Tratar los valores omitidos

Calcular el promedio de las columnas

```python
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

Reemplazar "NaN" por el valor medio en la columna "normalized-losses"

```python
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
avg_bore = df["bore"].astype("float").mean(axis=0)
print("Average of bore:", avg_bore)
```

Reemplazar "NaN" por el valor medio

```python
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

#### Pregunta #2.1: De acuerdo con el ejemplo anterior, reemplazar NaN en la columna "stroke" por la media

```python
avg_stroke = df["stroke"].astype("float").mean(axis=0)
print("Averange of stroke:", avg_stroke)
```

```python
df["stroke"].replace(np.nan, avg_stroke, inplace=True)
```

Calcular el valor medio de la columna "horsepower"

```python
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)
```

Reemplazar "NaN" por el valor medio

```python
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

Calcular el valor medio de la columna "peak-rpm"

```python
avg_peakrpm = df["peak-rpm"].astype("float").mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
```

Reemplazar "NaN" por el valor medio:

```python
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

#### Conceptos Fundamentales del Agrupamiento

```python
df['drive-wheels'].unique()
```

Si queremos saber, en promedio, qué tipo de rueda motriz es más valiosa, podemos agrupar
"drive- wheels" y luego calcular su promedio.

```python
df_group_one = df[['drive-wheels', 'body-style', 'price']]
```

A continuación, podemos calcular el precio medio de cada una de las distintas categorías de datos.

```python
df_group_one = df_group_one.groupby(['drive-wheels', 'body-style'], as_index=False).mean()
```

```python
df_group_one
```

```python
# grouping results
df_gptest = df[['drive-wheels', 'body-style','price']]
grouped_test1 = df_gptest.groupby(['drive-wheels','body-style'],as_index=False).mean()
grouped_test1
```

Dejaremos la variable "drive-wheel" como las filas de la tabla, y pivot "body- style" para que se convierta en las columnas de la tabla

```python
grouped_pivot = grouped_test1.pivot(index='drive-wheels',columns='body-style')
grouped_pivot
```

```python
grouped_pivot = grouped_pivot.fillna(0) #fill missing values with 0
grouped_pivot
```

Las etiquetas por defecto no nos ofrecen información útil. Cambiémoslo.

```python
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')
#label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index
#move ticks and labels to the center
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)
#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)
#rotate label if too long
plt.xticks(rotation=90)
fig.colorbar(im)
plt.show()
```
