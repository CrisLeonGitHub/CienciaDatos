# Regresi√≥n lineal con una variable



```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import collections
```

```python
def error(m, b, points):
    totalError = 0
    for i in range(0, len(points)):
        totalError += (points[i].y - (m * points[i].x + b)) ** 2
    return totalError / float(2*len(points))
```

```python
x1=np.array([30, 28, 35, 38, 100]);
y1=np.array([91,95,101,130,281]);
print(x1)
print(y1)

n=len(x1)
A=np.array([[x1[j], 1] for j in range(n)]) 
print(A)

B=y1
print(B)
```

```python
X=np.linalg.lstsq(A,B)[0]
print(X)
a=X[0]
b=X[1]
print(a)
print(b)

a1=a
a0=b
print(a1)
print(a0)
```

```python
plt.plot(x1,y1,'b*')
plt.plot(x1,a1*x1+a0,'r-')
plt.show()

y2=a1*x1+a0
print("salida estimada")
print(y2)

S=sum((y2-y1)*(y2-y1))
print(S)
```

```python
#code para 3d print
x = x1
y = y1
Point = collections.namedtuple('Point', ['x', 'y'])
points=[Point(x,y) for x,y in zip(x,y)]
#points = [Point(x,y)]
print(points)
print('Prueba 001')
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ms = np.linspace(-10, 10, 20)
bs = np.linspace(-60, 60, 20)

M, B = np.meshgrid(ms, bs)
zs = np.array([error(mp, bp, points) 
               for mp, bp in zip(np.ravel(M), np.ravel(B))])
print(zs.shape)
print(M.shape)    
Z = zs.reshape(M.shape)
print(M)
print(B)
print(Z)
ax.plot_surface(M, B, Z, rstride=1, cstride=1, color='b', alpha=0.5)

ax.set_xlabel('m')
ax.set_ylabel('b')
ax.set_zlabel('error')
plt.show()

#plt.figure()
#cs=plt.contour(M,B,Z)
#plt.show
#show(plt)
#m=2
#sigma=sqrt(S/(n-m))
#print(sigma)
```
