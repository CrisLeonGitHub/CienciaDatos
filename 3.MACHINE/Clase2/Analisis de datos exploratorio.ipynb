## 8. Detectando Outliers

Un valor atípico es un punto o conjunto de puntos que son diferentes de otros puntos. A veces pueden ser muy altas o muy bajas. A menudo es una buena idea detectar y eliminar los valores atípicos. Porque los valores atípicos son una de las principales razones para dar como resultado un modelo menos preciso. Por lo tanto, es una buena idea eliminarlos. La detección y eliminación de valores atípicos que voy a realizar se llama técnica de puntaje IQR. A menudo, los valores atípicos se pueden ver con visualizaciones utilizando un diagrama de caja. A continuación se muestra el diagrama de caja de MSRP, Cylinders, Horsepower y EngineSize. Aquí, en todas las parcelas, puede encontrar que algunos puntos están fuera de la caja, no son otros que valores atípicos. La técnica de encontrar y eliminar valores atípicos que estoy realizando en esta tarea se basa en un tutorial de [hacia la ciencia de datos] (https://towardsdatascience.com/ways-to-detect-and-remove-the-outliers-404d16608dba )

```python
sns.boxplot(x=df['Price'])
```

```python
sns.boxplot(x=df['HP'])
```

```python
sns.boxplot(x=df['Cylinders'])
```

```python
#detectando y eliminando outliers
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

No se preocupe por los valores anteriores porque no es importante conocer todos y cada uno de ellos porque es importante saber cómo usar esta técnica para eliminar los valores atípicos.

```python
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

Como se ve arriba, había alrededor de 1600 filas de valores atípicos. Pero no puede eliminar por completo los valores atípicos porque incluso después de usar la técnica anterior, puede haber 1 a 2 valores atípicos sin eliminar, pero eso está bien porque había más de 100 valores atípicos. Algo es mejor que nada.

## 9. Ploteo de diferentes características entre sí (dispersión), contra frecuencia (histograma)

### Histograma

El histograma se refiere a la frecuencia de aparición de variables en un intervalo. En este caso, existen principalmente 10 tipos diferentes de empresas de fabricación de automóviles, pero a menudo es importante saber quién tiene la mayor cantidad de automóviles. Hacer este histograma es una de las soluciones triviales que nos permite conocer el número total de automóviles fabricados por una compañía diferente.

```python
df.Make.value_counts().nlargest(100).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

### Heat Maps

Mapas de calor es un tipo de diagrama que es necesario cuando necesitamos encontrar las variables dependientes. Una de las mejores formas de encontrar la relación entre las características se puede hacer usando mapas de calor. En el siguiente mapa de calor sabemos que la característica del precio depende principalmente del tamaño del motor, la potencia y los cilindros.

```python
plt.figure(figsize=(15,10))
c= df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
```

## Scatterplot unidimensional 

Ahora veremos un scatter plot unidimensional. Lo usaremos para analizar visualmente el patron de una variable continua. Primero lo haremos de forma regular:

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df.index, df['Price'])
ax.set_xlabel('index') #index tiene los indices que por defecto tiene el objeto pandas 'df'. Para ver ello, pueden imprimirlo "print(df)". La columna sin rotulo con numeros es el 'index' de df.
ax.set_ylabel('Price')
plt.show()
```

Sin embargo este grafico lo podemos mejorar considerando un plot unidimensional ordenando la variable.

```python
df1=df.sort_values('Price')
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df.index, df1['Price'])
ax.set_xlabel('index')
ax.set_ylabel('Price')
plt.show()
```

### Scatterplot bidimensional

Generalmente usamos diagramas de dispersión para encontrar la correlación entre dos variables. Aquí las gráficas de dispersión se trazan entre Caballos de fuerza y ​​Precio y podemos ver la gráfica a continuación. Con la gráfica que se muestra a continuación, podemos dibujar fácilmente una línea de tendencia. Estas características proporcionan una buena dispersión de puntos.

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['HP'], df['Price'])
ax.set_xlabel('HP')
ax.set_ylabel('Price')
plt.show()
```
