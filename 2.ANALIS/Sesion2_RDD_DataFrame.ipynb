#2.-Crear un RDD de un conjunto de archivos  

##Leer archivo de texto en RDD TextFile()
El método se utiliza para leer un archivo de texto, el método toma la ruta como argumento y, opcionalmente, toma varias particiones como segundo argumento.

```python
# Lee los archivos que se encuentran carpeta y los almacena en variable text
src='./sample_data/Texto/*.txt'
rdd = sc.textFile(src)
```

```python
src='./sample_data/texto1.txt'
rdd1 = sc.textFile(src)
```

```python
src='./sample_data/texto2.txt'
rdd2 = sc.textFile(src)
```

```python
rdd1.distinct().collect()
```

##Collect
Acción que devolverá todos los elementos para el RDD dado.


```python
rdd.collect()
```

##Filter
Transformación que permite filtrar las entradas de valor y dejarla en otro RDD

```python
rddF = rdd.filter(lambda x : 'Big Data' in x)
rddF.collect()
```

##map
Transformación que mapeará su entrada a alguna salida basada en la función especificada en la función

```python
rddM= rdd.map(lambda x: x.upper())
rddM.collect()[0:10]
```

##flatMap: 
Transformación similar a map, pero puede devolver múltiples elementos para cada entrada en el RDD dado.
###split: Divide una cadena en una lista, donde cada palabra es un elemento de la lista

```python
# separa por palabras (split) con una palabra por registro
palabras = rdd.flatMap(lambda x: x.split())

# Se imprimen las primeras 10 palabras
palabras.collect()[0:10]
```

##map
Esto mapeará su entrada a alguna salida basada en la función especificada en la función


```python
# Genera las parejas <clave, valor> representandolas como la tupla (word, 1)
wc = palabras.map(lambda x: (x,1))
wc.collect()[0:10]
```

##reduceByKey
* Esta función reduce los pares de valores clave en función de las claves y una función determinada dentro de reduceByKey

```python
# Suma los valores para la misma clave. Spark internamente ordena por claves
counts = wc.reduceByKey(lambda x,y:x+y)
counts.collect()[0:10]
```

##sortByKey
Esta función realizará la clasificación en un par (clave, valor) RDD basado en las claves. De forma predeterminada, la clasificación se realizará en orden ascendente.
Para clasificar en orden descendente, pase  “ascending=False”.

```python
countsr = counts.map(lambda x: (x[1],x[0])).sortByKey(ascending=False)
#Print rdd6 result to console
print(countsr.collect())
```

##saveAsTextFile
Permite almacenar los datos en un archivo de salida

```python
countsr.saveAsTextFile("./sample_data/texto/output2")
```

#3.- Crear DataFrame

##Crea un DataFrame a partir de un archivo CSV
Podemos crear un DataFrame usando un archivo CSV y podemos especificar varias opciones como un separador, encabezado, esquema, inferSchema y varias otras opciones.

```python
archivo = './sample_data/Titanic.csv'
df= spark.read.csv(archivo,sep=',', inferSchema=True, header=True)
# imprimir tipo de archivo
print(type(df))
```

###Impotar Librerias de Funciones Sql
Para utilizar funciones agrupadas se debe importar las librerias de Sql

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
```

###Aggregations
Podemos usar la función groupBy para agrupar los datos y luego usar la función "agg" para realizar la agregación de datos agrupados.

```python
(df.groupBy("Sex")
  .agg(
      count("Age").alias("count"),
      sum("Age").alias("sum"),
      max("Age").alias("max"),
      min("Age").alias("min"),
      round(avg("Age"),1).alias("avg")
        ).show()
)
```

###Sorting
Ordena los datos según el "Fare". De forma predeterminada, la clasificación se realizará en orden ascendente.

```python
df.sort("Fare").show(5)
```

```python
df.sort(desc("Age")).show(5)
```

###Columnas derivadas
Podemos usar la función "withColumn" para derivar la columna en función de las columnas existentes ...

```python
df.withColumn("Meses", col("Age") * 12).show()
```
