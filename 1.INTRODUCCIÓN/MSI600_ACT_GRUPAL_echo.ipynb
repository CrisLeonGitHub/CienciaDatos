1. Para cada productor determine la cantidad de días promedio de fermentación.

```python
import pandas as pd
df = pd.read_excel('datos_vinos.xlsx',sheet_name=["1","2","3"]) #por defecto trae la hoja 1 que es igual a la 0


dfFermentaciones = df["1"]
dfFermentaciones.head()

dfProductores = df["2"]
dfProductores.head()


dfMuestras = df["3"]
dfMuestras.head()


dfjoin = pd.merge(dfFermentaciones,dfProductores,left_on="Fermentación",right_on="numero_ferm",how="outer") #asociamos columnas con nombre distinto


dfjoin2 = pd.merge(dfjoin,dfMuestras,left_on="Productor",right_on="Productor",how="outer") #asociamos columnas con nombre distinto

print("El promedio de fermentación por productor es :\n",dfjoin2.groupby("Productor")["Días Fermentación"].mean())

```

2. Si una fermentación finaliza con una densidad menor a 1000 y su ácidez volátil es menor a 1, entonces es considerada exitosa, de lo contrario es considerada problemática. Agregue una columna que permita determinar si una fermentación es exitosa o problemática. 

```python
#Respuesta 2
#funcion que valida estado de fermentacion
def fncEstadoFermentacion(densidad,acidez):
  
  if densidad<1000 and acidez<1:
    return "Exitosa"
  else:
      return "Problematica"

dfjoin2["Estado"] =dfjoin2.apply(lambda x: fncEstadoFermentacion(x["Densidad final"],x["Acidez Volátil"]),axis=1)

dfjoin2
```

3. Determine la cantidad de muestras promedio para las fermentaciones exitosas y problemáticas. 

```python
#Respuesta 3

print("La cantidad de muestras promedio es :\n",dfjoin2.groupby("Estado")["numero_muestras"].mean())

```

4. Determine el productor que obtuvo más fermentaciones exitosas así como el que obtuvo más fermentaciones problemáticas.

```python
#Respuesta 4

dfExitoso= dfjoin2.query("Estado=='Exitosa'")
dfProblematico = dfjoin2.query("Estado=='Problematica'")


df4=dfExitoso.groupby(['Productor'])['Estado'].count()
df4 = df4.sort_values(axis=0,ascending=False)
print("Productor con mas fermentaciones exitosas ", df4.head(1))

df5=dfProblematico.groupby(['Productor'])['Estado'].count()
df5 = df5.sort_values(axis=0,ascending=False)
print("Productor con mas fermentaciones problematicas ", df5.head(1))

```
