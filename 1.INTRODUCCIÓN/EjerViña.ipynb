```python
import pandas as pd
df = pd.read_excel('datos-2.xlsx',sheet_name=["fermentaciones","viñas","productor"])

dfFermentaciones = df["fermentaciones"]
dfFermentaciones.head()

dfVinas = df["viñas"]
dfVinas.head()

dfProductor = df["productor"]
dfProductor.head()

df6 = pd.merge(dfFermentaciones,dfVinas,left_on="Fermentación",right_on="Id Fermentación",how="outer") #asociamos columnas con nombre distinto

df6.head()
```

```python
dfjoin = pd.merge(dfFermentaciones,dfVinas,left_on="Fermentación",right_on="Id Fermentación",how="outer") #asociamos columnas con nombre distinto

dfjoin.head()

dfjoin2 = pd.merge(dfjoin,dfProductor,left_on="Viña",right_on="Viña",how="outer") #asociamos columnas con nombre distinto
print("Días Fermentacion Promedio")
print(dfjoin2.groupby("Viña")["Días Fermentación"].mean())
print("Días Fermentacion desviación")
print(dfjoin2.groupby("Viña")["Días Fermentación"].std())

print("Densidad final Promedio")
print(dfjoin2.groupby("Viña")["Densidad final"].mean())
print("Densidad final desviación")
print(dfjoin2.groupby("Viña")["Densidad final"].std())
```

```python
def fncPDF(densidad,dias):
  return  densidad/(dias**2)

def fncFermentacion(densidadfinal,diastotales):
  if (densidadfinal/diastotales**2)  < 10:
    return "Problematico"
  else:
      return "Exitoso"

dfjoin2["Estado"] =dfjoin2.apply(lambda x: fncFermentacion(x["Densidad final"],x["Días Fermentación"]),axis=1)
dfjoin2["ValorFermentacion"] =dfjoin2.apply(lambda x: fncPDF(x["Densidad final"],x["Días Fermentación"]),axis=1)


dfjoin2
```

```python
dfexitosa= dfjoin2.query("Estado=='Exitoso'")

print("La densidad minima Exitoso por Estado:\n",dfexitosa.groupby("Estado")["ValorFermentacion"].min())

print("La densidad maxima Exitoso por Estado:\n",dfexitosa.groupby("Estado")["ValorFermentacion"].max())

```

```python
dfexitosa= dfjoin2.query("Estado=='Problematico'")

print("La densidad Problematico minima por Estado:\n",dfexitosa.groupby("Estado")["ValorFermentacion"].min())

print("La densidad Problematico maxima por Estado:\n",dfexitosa.groupby("Estado")["ValorFermentacion"].max())
```
