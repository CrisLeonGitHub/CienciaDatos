```python
df = pd.DataFrame({'Name': ['Ana', 'Diana', 'John', 'Peter', 'Alex'],
                   'Gender': ['female','female','male','male','male'],
                   'Age': [20, 25 , 30, 55, 17], 
                   'Weight': [60, 76, 90, 87, 82],
                   'Height': [1.65, 1.78, 1.85, 1.72, 1.79]})
df.groupby("Gender")["Age","Weight"].mean()
```

```python
def imc(w,h):
  return w/h**2

def estado(m):
  if m < 25:
    return "Normal"
  else:
    return "Sobrepeso"

df["IMC"] = df.apply(lambda x:imc(x["Weight"],x["Height"]),axis=1)
df["Estado"] = df.apply(lambda x:estado(x["IMC"]),axis=1)
df
```

```python
df.pivot_table(index="Gender",values="IMC",aggfunc=["min","max","std","count","sum"])
```

```python
df.to_csv("datos.csv")
df.to_excel("hola.xlsx")
```
